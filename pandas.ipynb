```python
#14. ¿Cómo puedes devolver las especies únicas de pingüinos de df_penguins? ¿Cuántas especies únicas hay?
species = df_penguins["species"].copy()
unique = species.fillna(0)
unique = unique.drop_duplicates()
nrows = unique.shape[0]
print(unique)
print(f"existen {nrows} unicas especies, {list(unique.values)}.")
```
